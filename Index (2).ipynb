```python
from __future__ import (absolute_import, division,
                        print_function, unicode_literals)
import warnings
import pandas as pd
import matplotlib.pyplot as plt
import csv
import random as rd


class Transport:
    def __init__(self, Name, Speed, Power, Wheels):
        self.Name = Name
        self.Speed = Speed
        self.Power = Power
        self.Wheels = Wheels


class Transport1:
    CARS = Transport('Cars', rd.randint(90, 200), rd.randint(50, 190), rd.randint(4, 10))
    Buses = Transport('Buses', rd.randint(100, 245), rd.randint(100, 200), rd.randint(4, 10))
    Bikes = Transport('Bikes', rd.randint(110, 210), rd.randint(50, 190), rd.randint(4, 10))
    Trucks = Transport('Truck', rd.randint(90, 200), rd.randint(50, 190), rd.randint(4, 10))


class Transport2(Transport1):
    Convertible = CARS = Transport('Convertible', rd.randint(90, 200), rd.randint(50, 190), rd.randint(4, 10))
    Estate_cars = CARS = Transport('Estate_cars', rd.randint(100, 245), rd.randint(100, 200), rd.randint(4, 10))
    Motor_bikes = Bikes = Transport('Motor_bikes', rd.randint(110, 210), rd.randint(50, 190), rd.randint(4, 10))
    Vans = CARS = Transport('Vans', rd.randint(90, 200), rd.randint(50, 190), rd.randint(4, 10))
    Sport_cars = CARS = Transport('Sport_cars', rd.randint(90, 200), rd.randint(50, 190), rd.randint(4, 10))
    Pick_ups = CARS = Transport('Pick_ups', rd.randint(100, 245), rd.randint(100, 200), rd.randint(4, 10))
    Pedal_bikes = Bikes = Transport('Pedal_bikes', rd.randint(110, 210), rd.randint(50, 190), rd.randint(4, 10))
    Heavy_trucks = Trucks = Transport('Heavy_trucks', rd.randint(90, 200), rd.randint(50, 190), rd.randint(4, 10))
    Medium_trucks = Trucks = Transport('Medium_trucks', rd.randint(15, 90), rd.randint(50, 100), rd.randint(4, 10))


ut = Transport1.CARS.Name, Transport1.Trucks.Name, Transport1.Buses.Name, Transport1.Bikes.Name, Transport2.Heavy_trucks.Name, Transport2.Motor_bikes.Name, Transport2.Pick_ups.Name, Transport2.Pedal_bikes.Name, Transport2.Convertible.Name, Transport2.Medium_trucks.Name, Transport2.Estate_cars.Name, Transport2.Sport_cars.Name, Transport2.Vans.Name

with open('POU.csv', 'w+', newline='') as f:
    er = ['Name', 'Speed', 'Power', 'Wheels']
    thewrite = csv.DictWriter(f, fieldnames=er)

    thewrite.writeheader()
    thewrite.writerow({'Name': Transport1.CARS.Name, 'Speed': Transport1.CARS.Speed, 'Power': Transport1.CARS.Power,
                       'Wheels': Transport1.CARS.Wheels})
    thewrite.writerow(
        {'Name': Transport1.Trucks.Name, 'Speed': Transport1.Trucks.Speed, 'Power': Transport1.Trucks.Power,
         'Wheels': Transport1.Trucks.Wheels})
    thewrite.writerow({'Name': Transport1.Buses.Name, 'Speed': Transport1.Buses.Speed, 'Power': Transport1.Buses.Power,
                       'Wheels': Transport1.Buses.Wheels})
    thewrite.writerow({'Name': Transport1.Bikes.Name, 'Speed': Transport1.Bikes.Speed, 'Power': Transport1.Bikes.Power,
                       'Wheels': Transport1.Bikes.Wheels})
    thewrite.writerow({'Name': Transport2.Heavy_trucks.Name,
                       'Speed': Transport2.Heavy_trucks.Speed,
                       'Power': Transport2.Heavy_trucks.Power,
                       'Wheels': Transport2.Heavy_trucks.Wheels})
    thewrite.writerow({'Name': Transport2.Motor_bikes.Name,
                       'Speed': Transport2.Motor_bikes.Speed,
                       'Power': Transport2.Motor_bikes.Power,
                       'Wheels': Transport2.Motor_bikes.Wheels})
    thewrite.writerow({'Name': Transport2.Sport_cars.Name,
                       'Speed': Transport2.Sport_cars.Speed,
                       'Power': Transport2.Sport_cars.Power,
                       'Wheels': Transport2.Sport_cars.Wheels})
    thewrite.writerow({'Name': Transport2.Vans.Name,
                       'Speed': Transport2.Vans.Speed,
                       'Power': Transport2.Vans.Power,
                       'Wheels': Transport2.Vans.Wheels})
    thewrite.writerow({'Name': Transport2.Medium_trucks.Name,
                       'Speed': Transport2.Medium_trucks.Speed,
                       'Power': Transport2.Medium_trucks.Power,
                       'Wheels': Transport2.Medium_trucks.Wheels})
    thewrite.writerow({'Name': Transport2.Convertible.Name,
                       'Speed': Transport2.Convertible.Speed,
                       'Power': Transport2.Convertible.Power,
                       'Wheels': Transport2.Convertible.Wheels})
    thewrite.writerow({'Name': Transport2.Motor_bikes.Name,
                       'Speed': Transport2.Motor_bikes.Speed,
                       'Power': Transport2.Motor_bikes.Power,
                       'Wheels': Transport2.Motor_bikes.Wheels})
    thewrite.writerow({'Name': Transport2.Pick_ups.Name,
                       'Speed': Transport2.Pick_ups.Speed,
                       'Power': Transport2.Pick_ups.Power,
                       'Wheels': Transport2.Pick_ups.Wheels})
    thewrite.writerow({'Name': Transport2.Pedal_bikes.Name,
                       'Speed': Transport2.Pedal_bikes.Speed,
                       'Power': Transport2.Pedal_bikes.Power,
                       'Wheels': Transport2.Pedal_bikes.Wheels})

df = pd.read_csv('POU.csv')
df['Power'].value_counts().plot.pie()

s = df['Power']
rt = ['Cars', 'Truck', 'Buses', 'Bikes', 'Heavy_trucks', 'Motor_bikes', 'Sport_cars', 'Vans', 'Medium_trucks', 'Convertible', 'Motor_bikes','Pick_ups', 'Pedal_bikes']
cols = ['darkorange', 'c', 'k', 'w', 'orange', 'maroon', 'tan', 'r', 'g', 'coral', 'olive', 'chartreuse', 'gold']

plt.pie(s, labels=rt, colors=cols, startangle=900, )

# plt.xlabel('x')
# plt.ylabel('y')
plt.title('Transport')
# plt.legend()
plt.show()

df = pd.read_csv('POU.csv')
df['Wheels'].plot.pie()

s = df['Wheels']
rt = ['Cars', 'Truck', 'Buses', 'Bikes', 'Heavy_trucks', 'Motor_bikes', 'Sport_cars', 'Vans', 'Medium_trucks', 'Convertible', 'Motor_bikes','Pick_ups', 'Pedal_bikes']
cols = ['darkorange', 'c', 'k', 'w', 'orange', 'maroon', 'tan', 'r', 'g', 'coral', 'olive', 'chartreuse', 'gold']

plt.pie(s, labels=rt, colors=cols, startangle=900, )

# plt.xlabel('x')
# plt.ylabel('y')
plt.title('Transport')
# plt.legend()
plt.show()

dfe = pd.read_csv('POU.csv')
dfe['Speed'].plot.pie()

se = dfe['Speed']
rt = ['Cars', 'Truck', 'Buses', 'Bikes', 'Heavy_trucks', 'Motor_bikes', 'Sport_cars', 'Vans', 'Medium_trucks', 'Convertible', 'Motor_bikes','Pick_ups', 'Pedal_bikes']
cols = ['darkorange', 'c', 'k', 'w', 'orange', 'maroon', 'tan', 'r', 'g', 'coral', 'olive', 'chartreuse', 'gold']

plt.pie(se,labels=rt, colors=cols, startangle=900, )

# plt.xlabel('x')
# plt.ylabel('y')
plt.title('Transport')
# plt.legend()
plt.show()

```
